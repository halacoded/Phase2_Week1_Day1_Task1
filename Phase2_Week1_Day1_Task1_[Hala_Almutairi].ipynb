###6 . Write a Python script that performs the following tasks:
  * Define a list of cities: `['Dubai', 'Kuwait', 'Riyadh', 'Doha', 'Amman', 'Baghdad']`.
  * Create a function get_pressure(city) that:
    * Takes a city name as an argument.
    * Constructs the API request URL using the provided city name and your API key.
    * Makes a request to the OpenWeatherMap API to fetch the weather data.
    * Extracts and returns the atmospheric pressure from the response.
  * Iterate over the list of cities and call the get_pressure(city) function for each city.
 * Store the results in a dictionary with city names as keys and their corresponding pressures as values.
  * Print the atmospheric pressure for each city in a readable format.
  * Create a bar chart to visualize the atmospheric pressures of the cities using Matplotlib. Ensure to label the axes and provide a title for the chart.

````python
#
import requests
cities = ['Dubai', 'Kuwait', 'Riyadh', 'Doha', 'Amman', 'Baghdad']
KEY="c42a5ce6aecd0756b8ba5e1c25d7ac79"

def get_pressure(city):
  URL = f'https://api.openweathermap.org/data/2.5/weather?q={city}&appid={KEY}'
  response = requests.get(URL)
  if response.status_code == 200:
        data = response.json()
        return data['main']['pressure']
pressures=[]
for city in cities:
    pressure = get_pressure(city)
    pressures.append(pressure)

dic={
    'City': cities,
    'Pressure': pressures
}
df = pd.DataFrame(dic)
df
````

````python
import matplotlib.pyplot as plt
plt.figure(figsize=(5, 5))
plt.pie(df.Pressure, autopct='%0.1f%%', labels=df.City, shadow=True, startangle=45)
plt.title('atmospheric pressures of the cities')
plt.show()
````

````python
#!pip install yfinance==0.2.59
````

````python
import yfinance as yf #index is data in yahoo

# Set the start and end dates for the historical data
start_date = '2020-01-01'
end_date = '2025-01-01'
# Define the list of assets to download data for
tickers = ['TSLA', 'AAPL', 'MSFT', 'AMZN', 'GOOGL']
data = yf.download(tickers, start=start_date, end=end_date)
close_data = data['Close']
````

````python
data
````

###8 Create line plots of the 'Close' prices for each company on the same chart, ensuring each line is differentiated by color and labeled appropriately

````python
plt.figure(figsize=(12, 6))
for ticker in tickers:
    plt.plot(close_data.index, close_data[ticker], label=ticker)

plt.title('Closing Prices Over Time')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.legend() #make color of each companey show
plt.grid()
plt.show()
````

###9 Create a box plot of the 'Close' prices for each company on the same chart, allowing for easy comparison of price distributions.

````python
plt.figure(figsize=(10, 5))
plt.boxplot(close_data)

plt.title('Close prices for each company')
plt.ylabel('Close Prices')
plt.show()
````
